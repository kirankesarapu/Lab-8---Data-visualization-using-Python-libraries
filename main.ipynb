```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_data.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['release_year'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Release Year')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df, palette='viridis')
plt.title('Number of Shows by Type')
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='magma')
plt.title('Top 10 Directors with Most Content')
plt.xlabel('Number of Shows')
plt.ylabel('Director')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['type'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title('Proportion of Movies and TV Shows')
plt.ylabel('')
plt.show()
```
